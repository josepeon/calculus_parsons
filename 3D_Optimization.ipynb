 $f(x) = x^2 - 9x$ on $[-4, 2]$




```python
def f(x): return x**2 - 9*x
x = np.linspace(-4, 2, 100)
plt.plot(x, f(x))
plt.grid();
```

```python
x_sym = sy.Symbol('x')
f_sym = x_sym**2 - 9*x_sym
fprime = sy.diff(f_sym, x_sym)
```

```python
critical = sy.solve(sy.Eq(fprime, 0), x_sym)
critical
```

1. $$f(x,y) = x^2y^2$$

```python
def f(x, y): return x**2*y**2. #define function
x = np.linspace(-3, 3, 100)    #define x
y = np.linspace(-3, 3, 100)    #define y
X, Y = np.meshgrid(x, y)       #create grid of points from x and y

ax = plt.axes(projection = '3d')
ax.plot_surface(X, Y, f(X, Y))
```

```python
plt.contour(X, Y, f(X, Y), levels = 100)
```

3. $$h(x,y) = 3x^2 - 2xy + y^2 - 8y$$

### The Second Derivative

In 2-Dimensions:

- $f'' > 0  $ minimum
- $f'' < 0  $ maximum
- $f'' = 0  $ indeterminate

In 3-Dimensions:

$$D = f_{xx}(x_0, y_0)f_{yy}(x_0, y_0) - (f_{xy}(x_0, y_0))^2$$

- $D > 0$ and $f_{xx}(x_0, y_0) > 0$ -- minimum
- $D > 0$ and $f_{xx}(x_0, y_0) < 0$ -- max
- $D < 0$ -- saddle point
- $D = 0$, test is inconclusive
