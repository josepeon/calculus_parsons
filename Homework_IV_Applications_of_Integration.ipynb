####1. Dijksterhuis, Atomists vs. Platonists
Dijksterhuis explains that Atomists saw the universe as made of tiny, indivisible particles moving in empty space, while the Platonic school viewed it as ordered by ideal, continuous forms and geometry. He argues this contrast helped shift science toward a mechanical worldview where nature was explained through measurable motion rather than abstract perfection. The difference matters because it marked the beginning of modern scientific thinking.

####2. Deleuze & Guattari, A Thousand Plateaus
Deleuze and Guattari describe both atomistic and Platonic views as limited: one reduces everything to isolated parts, and the other to rigid ideals. They instead suggest reality works more like a network of dynamic flows and connections. Their writing is dense but intentionally challenges traditional thinking, using creative language to push readers toward new ways of understanding structure and change.

####3. Binomial Distribution
A binomial distribution models the number of successes in a fixed number of independent trials, each with the same probability of success. It connects to summations because probabilities across all possible outcomes are added up to form the full distribution. For example, flipping a coin 5 times and getting exactly 3 heads

####Normal Distribution
The normal distribution describes continuous data that clusters around a mean, shaped like a bell curve. It connects to integrals because the total probability equals 1, which requires integration.

####Method of Moments
The method of moments estimates parameters of a probability distribution by matching sample moments to theoretical moments.

####Pappus and Projective Geometry
Pappus of Alexandria was a Greek mathematician who studied geometry and volume. His ideas influenced Desargues and Monge, who developed projective geometry, a system studying geometric properties preserved under projection like perspective in art. This relates to Deleuzes philosophical discussions about space, continuity, and perspective as ways of structuring thought and perception
