##Annotated Bibliography
####1. Garland, Michael, and Paul S. Heckbert. “Surface Simplification Using Quadric Error Metrics.” Proceedings of SIGGRAPH, 1997.
Garland and Heckbert introduce Quadric Error Metrics (QEM), one of the most influential optimization methods used in modern 3D mesh decimation. The technique formulates mesh simplification as a least-squares minimization problem, using calculus-based optimization to preserve geometric accuracy while reducing polygon count. Because QEM is the basis for tools like ZBrush Decimation Master and many retopology engines used in 3D scanning workflows, this paper is essential to my project. It connects directly to how calculus, specifically minimization, error estimation, and gradients, is used to simplify models for animation, real-time rendering, and 3D printing.
####2. Botsch, Mario, et al. Polygon Mesh Processing. CRC Press, 2010.
This textbook provides an in-depth overview of the mathematical foundations of polygon mesh operations, including smoothing, curvature estimation, and surface optimization. Many of these operations rely on differential calculus, such as the use of derivatives to compute curvature or Laplacians to guide smoothing algorithms. The text shows how mathematical tools shape the workflows used in applications like C4D, Blender, Unreal Engine, and 3D scanning cleanup. This source supports my project by explaining how calculus enables accurate mesh manipulation, reduces geometric noise, and guides optimization processes.
####3. Kazhdan, Michael, and Hugues Hoppe. “Screened Poisson Surface Reconstruction.” ACM Transactions on Graphics, 2013.
This article describes Poisson Surface Reconstruction, one of the most widely used techniques for turning point cloud data into continuous meshes. The method solves a Poisson equation, a central PDE in calculus, using gradients, divergences, and Laplacians to create smooth, watertight surfaces from noisy 3D scan data. This paper is directly relevant to my work with 3D scanning, where I often need to reconstruct models before decimation or printing. It provides the math behind smoothing and surface approximation, showing how calculus converts raw scanning information into usable geometry.
####4. Sorkine, Olga. “Laplacian Mesh Processing.” Eurographics State-of-the-Art Reports, 2005.
Sorkine’s work explains the role of the Laplacian operator in mesh smoothing, deformation, and editing. Laplacian methods allow designers to preserve local geometric detail while adjusting global structure, which is essential for animation, rigging, and preparing assets for game engines. This source supports my project by showing how calculus-based operators shape many of the digital tools used in modern 3D workflows, especially in tasks like retopology and mesh refinement.
####5. Turk, Greg. “Re-Tiling Polygonal Surfaces.” Proceedings of SIGGRAPH, 1992.
Turk’s paper presents an early but foundational method for remeshing surfaces using gradient-based optimization. The algorithm redistributes vertices to achieve uniform polygon density, which is crucial for clean topology in 3D modeling, scanning, and printing. This research introduces several mathematical ideas such as energy minimization and geometric error estimation that remain central in modern mesh optimization tools. It supports my project by connecting mesh cleanup and decimation to calculus-driven processes that balance error and performance.
