# Learning ODE Parameters from Data

````python
from scipy.integrate import solve_ivp, odeint
from scipy.optimize import curve_fit
import numpy as np

xmin,xmax = 1,500

def f(t, y, b):
    dydt= -b * y
    return dydt

def curve(t, b, y0):
    sol= solve_ivp(lambda t, y: f(t, y, b),
               [xmin, xmax], [y0], t_eval= t)
    return sol.y[0]


xdata = np.linspace(xmin, xmax, 25)
ydata = np.exp(-0.02*xdata)+0.02*np.random.randn(*xdata.shape)

y0 = 5
b= 0.005
p0 = [b,y0]

popt, pcov = curve_fit(curve, xdata, ydata, p0=p0)
b, y0 = popt

print(f"b={b}, y0 = {y0}")
````

### Fitting the Lotka-Volterra System with SciPy's `curve_fit`

The Lotka-Volterra equations are a classic model describing the population dynamics of predators and prey. The system is defined by two coupled, first-order nonlinear differential equations:

$$\frac{dx}{dt} = \alpha x - \beta xy$$
$$\frac{dy}{dt} = \delta xy - \gamma y$$

where:
* $x(t)$ is the number of prey at time $t$.
* $y(t)$ is the number of predators at time $t$.
* $\alpha$ is the natural growth rate of prey in the absence of predators.
* $\beta$ is the rate at which predators consume prey.
* $\delta$ is the rate at which predators reproduce per prey consumed.
* $\gamma$ is the natural death rate of predators in the absence of prey.

To fit this system to experimental data using `scipy.optimize.curve_fit`, you would follow a similar procedure as outlined before, with the `my_system` function now representing the Lotka-Volterra equations:

1.  **Define the Lotka-Volterra system function:**

    ```python
    def lotka_volterra(t, a, b, d, g):
        def derivatives(z, time):
            x, y = z
            dx_dt = a*x - b*x*y
            dy_dt = d*x*y - g*y
            return [dx_dt, dy_dt]
        return derivatives
    ```

2.  **Numerically solve the system:** Use `scipy.integrate.odeint` to solve the system for given parameter values and initial conditions.

    ```python
    from scipy.integrate import odeint
    import numpy as np

    # Time points of your data
    time_points = np.array([...])
    # Initial conditions for prey and predator populations
    initial_conditions = [x0, y0]

    def solve_lotka_volterra(params):
        a, b, d, g = params
        derivs = lotka_volterra(None, a, b, d, g)
        solution = odeint(derivs, initial_conditions, time_points)
        return solution[:, 0], solution[:, 1] # Return prey and predator populations
    ```

3.  **Define the fitting function:** This function will take the time points and the parameters as input and return the simulated prey and/or predator populations at those time points.

    ```python
    def fit_lotka_volterra(t, a, b, d, g, x0, y0):
        global initial_conditions
        initial_conditions = [x0, y0]
        prey_sim, predator_sim = solve_lotka_volterra([a, b, d, g])
        # You might choose to fit to both prey and predator data or just one.
        # Here, we'll return both, and curve_fit can handle multiple output arrays.
        return np.concatenate((prey_sim, predator_sim))
    ```

4.  **Prepare your experimental data:** You should have time series data for both the prey and predator populations.

    ```python

    prey_data = np.array([...])
    predator_data = np.array([...])
    combined_data = np.concatenate((prey_data, predator_data))
    ```

5.  **Use `scipy.optimize.curve_fit`:**

    ```python

    import matplotlib.pyplot as plt

    # Initial guess for the parameters [alpha, beta, delta, gamma, initial_prey, initial_predator]
    initial_guess = [...]

    popt, pcov = curve_fit(fit_lotka_volterra, time_data, combined_data, p0=initial_guess)

    # popt will contain the optimized parameters: [alpha, beta, delta, gamma, x0, y0]
    a_opt, b_opt, d_opt, g_opt, x0_opt, y0_opt = popt

    initial_conditions_fitted = [x0_opt, y0_opt]
    derivs_fitted = lotka_volterra(None, a_opt, b_opt, d_opt, g_opt)
    fitted_solution = odeint(derivs_fitted, initial_conditions_fitted, time_data)
    fitted_prey = fitted_solution[:, 0]
    ```

**Important Considerations for Fitting Lotka-Volterra:**

* **Parameter Sensitivity:** The Lotka-Volterra model can be sensitive to the initial parameter guesses. Providing reasonable initial values based on biological intuition or prior knowledge can be crucial for successful fitting.
* **Data Quality:** The quality and quantity of your experimental data will significantly impact the accuracy of the fitted parameters.
* **Model Limitations:** The basic Lotka-Volterra model is a simplification of real ecological interactions. It might not capture all the complexities of your system. Consider extensions of the model if necessary.
* **Error Handling:** Be prepared for potential errors during the optimization process (e.g., the algorithm not converging).

Suppose we know the concentration of A follows this differential equation:

$$\frac{dC_A}{dt}=kCA$$

and we have data we want to fit to it. Here is an example of doing that.

````python
def lotka_volterra(Variables, t, Alpha, Beta, Gamma, Delta):
  drdt = Variables[0]*(Alpha - Beta*Variables[1])
  dwdt = -Variables[1]*(Gamma - Delta*Variables[0])
  return [drdt, dwdt]
````

````python
def sir(variables, t, beta, gamma):
  S, I = variables[0], variables[1]
  dsdt = -beta*S*I
  didt = beta*S*I - gamma*I
  drdt = gamma*I
  return [dsdt, didt, drdt]
````

````python
variables = [762, 1, 0] #susceptible, infected
beta = .00218
gamma = .44
````

````python
t = np.linspace(0, 20, 21)
````

````python
solns = odeint(sir, variables, t, args = (beta, gamma,))
````

````python
plt.plot(t, solns[:, 0], label = 'susceptible')
plt.plot(t, solns[:, 1], label = 'infected')
plt.plot(t, solns[:, 2], label = 'recovered')
plt.grid()
plt.legend();
````
