### Differential Equations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, 20)
y = np.linspace(-2, 4, 20)
X, Y = np.meshgrid(x, y)
dx = .5*np.sin(np.pi/2*X)*np.sqrt(Y+7)
dy = np.ones(dx.shape)
plt.quiver(X, Y, dx, dy)
```

```python
t = np.arange(2, 3, .1)
```

```python
from scipy.integrate import odeint
```

```python
def dydx(y, x):
  return .5*np.sin(np.pi/2*x)*np.sqrt(y+7)
```

```python
solns = odeint(dydx,y0 = 1, t=t)
```

```python
x = np.linspace(-3, 3, 20)
y = np.linspace(-2, 4, 20)
X, Y = np.meshgrid(x, y)
dx = .5*np.sin(np.pi/2*X)*np.sqrt(Y+7)
dy = np.ones(dx.shape)
plt.quiver(X, Y, dx, dy)
plt.plot(t, solns, '-r')
```

```python
x = np.linspace(0, 10, 11)
y = np.linspace(0, 25, 21)
X, Y = np.meshgrid(x, y)
```

```python
dx = .1*X
dy = np.ones(dx.shape)
dx = dx/(dx**2 + dy**2)**0.5
dy = dy/(dx**2 + dy**2)**0.5
plt.quiver(X, Y, dx, dy)
plt.plot(t, solns)
plt.ylim(0, 25)
```
