### Max/Min Problems in 3-Dimensions

For each of the problems, determine the partial derivatives of the function, the critical points, and decide if that is a maximum or minimum value.

Solve problems 319 - 331 odd from OpenStax [here](https://openstax.org/books/calculus-volume-3/pages/4-7-maxima-minima-problems).

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sympy as sy
from mpl_toolkits import mplot3d
```

$$319.	ùëì(ùë•,ùë¶)=ùë•^2ùë¶^2$$

```python
x, y = sy.symbols('x y')
f = x**2 * y**2
fx = sy.diff(f, x)
fy = sy.diff(f, y)
fxx = sy.diff(fx, x); fyy = sy.diff(fy, y); fxy = sy.diff(fx, y)

fx, fy, fxx, fyy, fxy
```

```python
crit = sy.solve([fx, fy], (x, y))
crit
```

$$321.	 ùëì(ùë•,ùë¶)=2ùë•ùë¶+3ùë•+4ùë¶$$

```python
x, y = sy.symbols('x y')
f = 2*x*y + 3*x + 4*y
fx, fy = sy.diff(f,x), sy.diff(f,y)
fxx, fyy, fxy = sy.diff(fx,x), sy.diff(fy,y), sy.diff(fx,y)
crit = sy.solve([fx, fy], (x, y))
fx, fy, fxx, fyy, fxy, crit
```

$$327.	 ùëì(ùë•,ùë¶)=3ùë•^2‚àí2ùë•^ùë¶+ùë¶^2‚àí8ùë¶$$

$$329.	 ùëì(ùë•,ùë¶)=ùë¶^2+ùë•ùë¶+3ùë¶+2ùë•+3$$
